```python
codon_table=open('C:/Users/mhyavas/Desktop/Rosalind-Bioinformatic Stringhold/codon_table.txt','r')
rna_string=open('C:/Users/mhyavas/Downloads/rosalind_prot(3).txt','r')
codon_line=codon_table.read().split()
rna_line=rna_string.read().strip()
codon_dict={}
codon_dict={codon_line[i]:codon_line[i+1] for i in range(0,len(codon_line),2)}
letter_protein=[]
```

```python
for i in range(0,len(rna_line),3):
    if rna_line[i:i+3] in codon_dict.keys():
        letter_protein.append(codon_dict[rna_line[i:i+3]])
letter_protein.remove('Stop')
print(''.join(letter_protein))
```
